```python
import random

#cards variable
suits = ["clubs", "spades", "hearts","diamonds"]
faces = ["Jack","Queen", "King", "Ace"]
numbered = [2,3,4,5,6,7,8,9,10]

#introducing "set" data structure
deck = set()
for suit in suits :
    for card in (faces + numbered):
        deck.add((card,"of",suit))

#randomly choose card from the entire set "deck" of length 52
def draw() :
     card = random.choice(list(deck))
     return card

#print card
for _ in range(len(deck)):
    print(draw())

#check if any card exists
cardToCheck = ("Ace","of","hearts")

if cardToCheck in deck:
    print("Exists")
else :
    print ("Not Exists")
```
