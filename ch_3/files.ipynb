```python
import swimclub
import os

swimFiles = os.listdir(swimclub.FOLDER)

if len(swimFiles) > 0 :

    swimFiles.remove(".DS_Store")

    for count, fileName in enumerate(swimFiles,1) :
        print (count, "Processing : " , fileName)
        print("----------------------------")
        swimData = swimclub.readSwimData(fileName)
        swimmer, ageGroup, distance, stroke,times,convertedAverageTime = swimData
        print("Swimmer - ", swimmer)
        print("Age group - ", ageGroup)
        print("Distance - ", distance)
        print("Stroke - ", stroke)
        print("Times - ", times)
        print("Average - ",convertedAverageTime)


else : 
    print ("No File Found")
```
